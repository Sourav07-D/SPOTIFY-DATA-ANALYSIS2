```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df_tracks.set_index("release_date",inplace=True)
df_tracks
```

```python

df_tracks.index=pd.to_datetime(df_tracks.index,format='ISO8601')
df_tracks.head()
```

```python
df_tracks[['artists']].iloc[10]
```

```python
df_tracks['artists'].iloc[10]
```

```python
df_tracks['duration_s']=df_tracks['duration_ms'].apply(lambda x:round(x/1000))
```

```python
df_tracks.drop('duration_ms',axis=1,inplace=True)
```

```python
df_tracks.head()
```

```python
df_tracks['duration_m']=df_tracks['duration_s'].apply(lambda x:x/60)
```

```python
df_tracks.head()
```

```python
df_tracks.duration_m.head()
```

```python
corr_df = df_tracks.drop(['id', 'key', 'mode', 'explicit','name','artists','id_artists'], axis=1).corr(method='pearson')

plt.figure(figsize=(14, 6))
heatmap = sns.heatmap(corr_df, annot=True, fmt='.1g', vmin=-1, vmax=1, center=0, cmap='inferno', linewidths=1, linecolor='Black')
heatmap.set_title("Correlation Heatmap Between Variables")
heatmap.set_xticklabels(heatmap.get_xticklabels(), rotation=90)
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='loudness',x='energy',color='c').set(title='Loudness vs Energy Corr')
```

```python
plt.figure(figsize=(10,6))
sns.regplot(data=sample_df,y='popularity',x='acousticness',color='b').set(title='Popularity vs Acousticness Corr')
```

```python
df_tracks['dates']=df_tracks.index.get_level_values('release_date')
df_tracks.dates=pd.to_datetime(df_tracks.dates)
years=df_tracks.dates.dt.year
```

```python
sns.displot(years,discrete=True,aspect=2,height=5,kind='hist').set(title="Number Of Songs Per Year")
```

```python
total_dr=df_tracks.duration_s
fig_dims=(18,7)
fig, ax=plt.subplots(figsize = fig_dims)
fig=sns.barplot(x=years,y=total_dr,ax=ax,errwidth=False).set(title="Year vs Duration")
plt.xticks(rotation=90)
```

```python
total_dr=df_tracks.duration_s
sns.set_style(style='whitegrid')
fig_dims=(18,7)
fig, ax=plt.subplots(figsize = fig_dims)
fig=sns.lineplot(x=years,y=total_dr,ax=ax).set(title="Year vs Duration")
plt.xticks(rotation=60)
```

```python
df_genre=pd.read_csv('SpotifyFeatures[1].csv')
```

```python
df_genre.head()
```

```python
plt.title('Duration Of Songs In Different Genres')
sns.color_palette('rocket', as_cmap=True)
sns.barplot(data=df_genre,x='duration_ms',y='genre')
plt.xlabel('Duration in MiliSeconds')
plt.ylabel('Genres')
```
